```python
#import libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# prepare some data
x = [10, 20, 30, 40, 50]
y = [3, 5, 10, 3, -4]

# Tell python to render plot in notebook 
output_notebook()

# Set up line graph
p = figure(title = "line example", x_axis_label = 'x', y_axis_label = 'y')

# add a line renderer with legend and line thickness
p.line(x, y, legend = "Random Number", line_width = 3)

# show the results
show(p)
```
